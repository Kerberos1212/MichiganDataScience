```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA 
%matplotlib notebook
```

```python

def main():
    df = pd.read_csv('food-consumption.csv', header = 0, sep = ',')
    X = df.iloc[:, 1:].values 
    print('----shape of training data------')
    print(X.shape)
    sc = StandardScaler()  
    X = sc.fit_transform(X) 
    pca = PCA(n_components=2) 


    X_pca = pca.fit_transform(X) 


    print('----shape of mean matrix------')
    print(pca.mean_.shape)

    print('----shape of eigenvectors matrix-----')
    print(pca.components_)
    print(pca.components_.shape)
    #plt.plot(np.transpose(pca.components_))
    print('---shape of eigenvalues matrix----')
    print(pca.singular_values_)
    print(pca.singular_values_.shape)

    print('---explained variance ratio -----------')
    explained_variance = pca.explained_variance_ratio_ 
    print(explained_variance)

    print('---transformed data--------------------')
    transformed_data = pd.DataFrame(data = X_pca
             , columns = ['principal component 1', 'principal component 2'])
    transformed_data['Country'] = df.iloc[:,0]
    print(transformed_data)
    print(transformed_data.shape)

    sns.scatterplot(data=transformed_data, x='principal component 1', y = 'principal component 2', hue='Country')
    plt.legend(loc='upper center',ncol=3)

    #sns.lmplot(data=transformed_data, x='principal component 1', y = 'principal component 2', hue='Country', fit_reg=False)
    #plt.scatter(transformed_data['principal component 1'], transformed_data['principal component 2'])
```

```python
main()
```
